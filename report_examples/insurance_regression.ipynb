# Practical Task 1: Simple Linear Regression
In this section, we load the insurance dataset to investigate the relationship 
between **AGE** (independent variable) and **Insurance Costs** (dependent 
[cite_start]variable).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the insurance dataset 
try:
    df_insurance = pd.read_csv('insurance.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: 'insurance.csv' not found.")
```

```python
# Assign variables 
X = df_insurance[['age']] # Independent variable
y = df_insurance['charges'] # Dependent variable

# Generate the first scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='dodgerblue', alpha=0.5)
plt.title('Insurance Costs vs Age (Initial Plot)')
plt.xlabel('Age')
plt.ylabel('Charges (R)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Initialise and fit the model using sklearn 
model = LinearRegression()
model.fit(X, y)

# Make predictions on the data to create the best-fit line 
y_pred = model.predict(X)

print(f"Model Training Complete.")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope (Coefficient for Age): {model.coef_[0]:.2f}")
```

```python
# Generate the second scatter plot with the regression line 
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='dodgerblue', alpha=0.4, label='Actual Data')
plt.plot(X, y_pred, color='crimson', linewidth=2, label='Best-Fit Line')

plt.title('Insurance Costs vs Age (With Regression Line)')
plt.xlabel('Age')
plt.ylabel('Charges (R)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### **Conclusion and Model Notes**

* **Observation**: The scatter plots show a clear upward trend, indicating that 
    as age increases, insurance charges also tend to rise .
* **Linear Regression**: The "Best-Fit Line" represents the model's prediction 
    of the average cost for any given age.
